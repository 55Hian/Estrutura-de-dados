1) Este exercício envolve a implementação dos métodos **insert_element**, **remove_element**, **search_element** e **reverse_array** na classe **ArrayManipulator**. Encorajo você a preencher o código desses métodos para realizar as operações correspondentes nos arrays.


```python
class ArrayManipulator:
    def __init__(self, array):
        self.array = array

    def get_array(self):
        return self.array

    def insert_element(self, index, element):
        
        """
        Insere um elemento na posição especificada do array.

        Parameters:
        - index (int): Índice onde o elemento será inserido.
        - element: Elemento a ser inserido.

        Retorna:
        - True se a inserção for bem-sucedida, False caso contrário.
        """
        # Implemente o código aqui
        
        # self array_exemplo.append(element)
        novaArray = self.array
        parteUmArray = []
        parteDoisArray = []
        elementoArray = [element]

        for x in range(len(novaArray)):
            if( x < index):
                parteUmArray.append(novaArray[x])
            else:
                parteDoisArray.append(novaArray[x])
        
        self.array = parteUmArray + elementoArray + parteDoisArray

        return True

    def remove_element(self, element):
        """
        Remove a primeira ocorrência do elemento no array.

        Parameters:
        - element: Elemento a ser removido.

        Retorna:
        - True se a remoção for bem-sucedida, False caso contrário.
        """
        # Implemente o código aqui
        novaArray = self.array

        for x in novaArray: 
            if( x == element):
                novaArray.remove(x)
                self.array = novaArray
                return True
        return False
        


    def search_element(self, element):
        
        
        """
        Procura a primeira ocorrência do elemento no array.

        Parameters:
        - element: Elemento a ser procurado.

        Retorna:
        - O índice do elemento se encontrado, ou -1 caso contrário.
        """
        # Implemente o código aqui
        teste = self.array

        
        for x in range(len(teste)):
            if teste[x] == element:
                return x
            
        return -1        
            
         

    def reverse_array(self):
        """
        Inverte a ordem dos elementos no array.

        Retorna:
        - O array invertido.
        """
        # Implemente o código aqui

        novaArray = self.array
        arrayReverse = []

        for x in range(len(novaArray)-1, -1, -1):
            arrayReverse.append(novaArray[x])

        self.array = arrayReverse

        return self.array 


# Exemplo de uso:
array_exemplo = [52, 43, 35, 22, 54]

manipulador = ArrayManipulator(array_exemplo)

# Teste dos métodos
print("Array original:", manipulador.get_array())

# Insira um elemento na posição 2
manipulador.insert_element(3, 66)
print("Array após inserção:", manipulador.get_array())

# Remova o elemento 4
manipulador.remove_element(52)
print("Array após remoção:", manipulador.get_array())

# Procure o elemento 3 - ok
indice = manipulador.search_element(52)
print("Índice do elemento 3:", indice)

# Inverta o array
array_invertido = manipulador.reverse_array()
print("Array invertido:", array_invertido)
```
